```python
import pandas as pd
import numpy as np

data_noise = pd.read_csv("../Data/hourly_noisedata_2022.csv")
data_noise.head()

data_Namen = data_noise[data_noise["description"]=="MP 03: Naamsestraat 62 Taste"]
print(data_Namen.head())

time = np.array(range(len(data_Namen)))
print(type(time))
```

```python
import plotly.graph_objects as go

# Create a line plot using Plotly
fig = go.Figure(data=go.Scatter(x=time, y=data_Namen["laeq"], mode='lines'))

# Customize the plot layout
fig.update_layout(
    #title="Noise Levels Over Time - MP 03: Naamsestraat 62 Taste",
    #xaxis_title="Time",
    #yaxis_title="Laeq",
    margin=dict(l=0, r=0, t=0, b=0),
    plot_bgcolor='rgba(0, 0, 0, 0)',
    height = 150,
    width = 1200
)

# Display the plot
fig.show()
```
