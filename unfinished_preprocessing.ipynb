```python
# conbine the meteo dataset together
meteocombined = pd.concat([Q1_2022, Q2_2022, Q3_2022, Q4_2022], axis=0)
meteocombined.head()
```

```python
# aggregate meteo data by day
avg_meteo_combined = meteocombined.groupby(['Year','Month', 'Day']).mean()
avg_meteo_combined = avg_meteo_combined.reset_index()

avg_meteo_combined.head()
```

```python
# conbine noise level noise42 together
noise42_combined = pd.concat([exp42_naamse35, exp42_naamse57, exp42_naamse62, exp42_calvarie, exp42_naamse81, exp42_park, exp42_kiosk, exp42_vrijt, exp42_his], axis=0)
noise42_combined.head()
```

```python
# Extract the date, month, hour, minute of "result_timestamp"
noise42_combined['result_timestamp'] = pd.to_datetime(noise42_combined['result_timestamp'], format='%d/%m/%Y %H:%M:%S.%f')


noise42_combined['result_date'] = noise42_combined['result_timestamp'].dt.date
noise42_combined['result_month'] = noise42_combined['result_timestamp'].dt.month
noise42_combined['result_day'] = noise42_combined['result_timestamp'].dt.day
noise42_combined['result_hour'] = noise42_combined['result_timestamp'].dt.hour
noise42_combined['result_minute'] = noise42_combined['result_timestamp'].dt.minute

noise42_combined.head()
```

```python
# aggregate noise42 by location & date
avg_noise42_combined = noise42_combined.groupby(['result_date','description']).mean()
avg_noise42_combined = avg_noise42_combined.reset_index()
columns_to_drop = ['result_hour', 'result_minute']
avg_noise42_combined.drop(columns_to_drop, axis=1, inplace=True)
avg_noise42_combined.head(150)
```
